# 1. Penerapan Bag of Words (BoW)

```python
# Import library CountVectorizer untuk metode Bag of Words
from sklearn.feature_extraction.text import CountVectorizer

# Dokumen teks yang akan diproses
documents = [
    "saya sedang mengerjakan project UTS Text Mining,"
    "yaitu Text Vectorizer dengan menerapkan Bag Of Words,"
    "TF-IDF, dan Word Embedding"
]

# Membuat objek CountVectorizer
vectorizer = CountVectorizer()

# Mengubah dokumen menjadi bentuk vektor BoW
vectors = vectorizer.fit_transform(documents)

# Menampilkan daftar kata unik (vocabulary)
print("Vocabulary:", vectorizer.get_feature_names_out())

# Menampilkan hasil vektor BoW dalam bentuk array
print("\nBoW Vector:")
print(vectors.toarray())
```

# 2. Penerapan TF-IDF

```python
# Import library TfidfVectorizer untuk metode TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

# Dokumen teks yang akan diproses
documents = [
    "saya sedang mengerjakan project UTS Text Mining,"
    "yaitu Text Vectorizer dengan menerapkan Bag Of Words,"
    "TF-IDF, dan Word Embedding"
]
# Membuat objek TfidfVectorizer
vectorizer = TfidfVectorizer()

# Mengubah dokumen menjadi bentuk vektor TF-IDF
vectors = vectorizer.fit_transform(documents)

# Menampilkan daftar kata unik (vocabulary)
print("Vocabulary:", vectorizer.get_feature_names_out())

# Menampilkan hasil vektor TF-IDF dalam bentuk array
print("\nTF-IDF Vector:")
print(vectors.toarray())
```

# 3. Penerapan Word Embedding

```python
# Import library Word2Vec dari gensim
from gensim.models import Word2Vec

# Data kalimat yang sudah dipisah per kata (tokenisasi manual)
sentences = [
    ["saya", "sedang", "mengerjakan", "project", "uts", "text",
     "mining", "yaitu", "text", "vectorizer","dengan", "menerapkan",
     "bag", "of", "words", "tf", "idf", "dan", "word", "embedding"]
]

# Membuat model Word2Vec
model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

# Menampilkan vektor embedding untuk kata "text"
print("Vector kata 'text':")
print(model.wv["text"])
```
